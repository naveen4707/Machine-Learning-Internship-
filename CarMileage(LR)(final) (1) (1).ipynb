```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Re-loading data and preparing input/output if kernel state is empty
# This makes the cell self-contained and runnable even if the kernel was reset
cardf = pd.read_csv("/root/.cache/kagglehub/datasets/trohitbharadwaj/car-mileage-prediction/versions/1/mtcars.csv")
inp = cardf[['wt']]
out = cardf['mpg']

# Re-initialize and fit the Linear Regression model
LR = LinearRegression()
LR.fit(inp, out)

# Predict for all input data
pred_all = LR.predict(inp) # Changed 'ind' to 'inp'

plt.figure(figsize=(10,6))
plt.scatter(inp, out) # Changed 'dep' to 'out'
plt.plot(inp,LR.predict(inp))
plt.xlabel("Actual Miles per Gallon (mpg)") # Corrected label
plt.ylabel("Predicted Miles per Gallon (mpg)") # Corrected label
plt.title("Actual vs Predicted Miles per Gallon (mpg) (Linear Regression Model)") # Corrected title
plt.grid(True)
plt.show()
```
