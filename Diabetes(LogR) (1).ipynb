```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dia=pd.read_csv("/content/drive/MyDrive/diabetcsv.csv")
dia.head()
```

```python
dia.isnull().sum()
```

```python
ind=dia[['age','mass','insu','plas']]
dep=dia['class']
```

```python
Logr=LogisticRegression()
Logr.fit(ind,dep)
```

```python
age=int(input("Enter Age:"))
mass=int(input("Enter mass:"))
insulin=int(input("Enter insulin:"))
plasma=int(input("Enter plasma:"))
pred=Logr.predict([[age,mass,insulin,plasma]])
print(pred)
```

```python
Logr.score(ind,dep)
```
