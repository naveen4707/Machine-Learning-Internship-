```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harrywang/wine-dataset-for-clustering")

print("Path to dataset files:", path)
```

```python
x=pd.read_csv("/root/.cache/kagglehub/datasets/harrywang/wine-dataset-for-clustering/versions/1/wine-clustering.csv")
print(x)
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(x)
labels=kmeans.labels_
print(labels)
```

```python
plt.scatter(x.iloc[:,6], x.iloc[:,12], c=labels)
plt.scatter(
    kmeans.cluster_centers_[:,6],
    kmeans.cluster_centers_[:,12],
    marker='x',
    s=200
)
plt.xlabel("Magnesium")
plt.ylabel("Proline")
plt.title("K-Means Clustering (K=4)")
plt.show()
```
