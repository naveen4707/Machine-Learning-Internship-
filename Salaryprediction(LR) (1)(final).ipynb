```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf=pd.read_csv('/content/drive/MyDrive/Salary Data.csv')
```

```python
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
saldf=saldf.dropna()
```

```python
inp=saldf[['Years of Experience']]
out=saldf['Salary']
```

```python
LR=LinearRegression()
LR.fit(inp,out)
```

```python
LR.predict([[5]])
```

```python
plt.figure()

# scatter plot (dataset)
plt.scatter(saldf['Years of Experience'], saldf['Salary'])

# regression line
plt.plot(saldf['Years of Experience'], LR.predict(inp))

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Experience vs Salary")

plt.show()
```
